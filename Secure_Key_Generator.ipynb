```python
import random
import string
import matplotlib.pyplot as plt

length = int(input("Enter password length: "))

characters = string.ascii_letters + string.digits
password = ""

upper = 0
lower = 0
digits = 0

for i in range(length):
    char = random.choice(characters)
    password += char

    if char.isupper():
        upper += 1
    elif char.islower():
        lower += 1
    elif char.isdigit():
        digits += 1

print("\nGenerated Password:", password)

# Strength check
if length < 6:
    print("Strength: Weak")
elif length < 10:
    print("Strength: Medium")
else:
    print("Strength: Strong")

# Graph
labels = ["Uppercase", "Lowercase", "Digits"]
values = [upper, lower, digits]

plt.bar(labels, values)
plt.title("Password Composition")
plt.xlabel("Character Types")
plt.ylabel("Count")
plt.show()
```


Objective:

To generate random secure keys using Python and analyze their strength.

Conclusion:

This project creates strong passwords and visually shows the character distribution.
